
In the Year 2022, how many houses (rows) located in the Greenwich Locality have exactly num_room == 3, and facing towards either the North or East?

```python
Criteria_Year = df.Year == 2022
```

```python
Criteria_Loc = df.Locality == 'Greenwich'
```

```python
Criteria_rooms = df.num_rooms == 3
```

```python
Criteria_Face = (df.Face == 'North') | (df.Face == 'East')
```

```python
Criteria = (df.Year == 2022) & (df.Locality == 'Greenwich') & (df.num_rooms == 3) & ((df.Face == 'North') | (df.Face == 'East'))
```

```python
Criteria = Criteria_Year & Criteria_Loc & Criteria_rooms & Criteria_Face
```

```python
df.loc[Criteria].shape
```

```python
df.loc[Criteria]
```
